```python
# Tarih sütununu tarih nesnelerine dönüştürme
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Tarih sütunundan yıl, ay ve gün özelliklerini çıkarma
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
addres = df["Address"]
addres = df['Address'].str.split(' ', expand=True)

df['No'] = addres[0]
df['Street'] = addres[1]
df['type_of_address'] = addres[2]
df = df.drop(columns=['Date',"Address"])
df
```

```python

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['Suburp_encoded'] = label_encoder.fit_transform(df['Suburb'])
df['Type_encoded'] = label_encoder.fit_transform(df['Type'])
df['Method_encoded'] = label_encoder.fit_transform(df['Method'])
df['SellerG_encoded'] = label_encoder.fit_transform(df['SellerG'])
df['Regionname_encoded'] = label_encoder.fit_transform(df['Regionname'])
df['CouncilArea_encoded'] = label_encoder.fit_transform(df['CouncilArea'])
df['Street_encoded'] = label_encoder.fit_transform(df['Street'])
df['type_of_address_encoded'] = label_encoder.fit_transform(df['type_of_address'])
```

```python

df = df[df['No'].notna()]
df
```

```python

no_column = df['No'].values
no_cleaned = [value.replace('/', '').replace('A', '').replace('B', '').replace('C', '') for value in no_column]

print(no_cleaned)
# Yeni bir "No_clean" sütunu oluşturun
df['No_clean'] = no_cleaned

df = df[df['Price'].notna()]
df['No_clean'] = pd.to_numeric(df['No_clean'], errors='coerce', downcast='integer')
df = df[df['No_clean'].notna()]
```

```python
df = df.drop(columns=['Suburb','Type','Method','SellerG','Regionname','CouncilArea','Street','type_of_address','No'])
df
y = df['Price']
X = df.drop("Price", axis=1)  # Özellikler
print(df.dtypes)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(y)
#Burada XGBoost modeli oluşturuyoruz, buradaki paramatreler değiştilerek belki daha başarılı sonuçlar elde edilebilir.
model = XGBRegressor(n_estimators=100, max_depth=7, random_state=42)
model.fit(X_train, y_train)
```
